# Restormer Video Restoration Benchmark

This notebook evaluates the performance of the Restormer architecture for video frame restoration on Apple Silicon.

Objectives:
- Restore blurred video frames
- Compare original vs restored outputs
- Measure processing time
- Evaluate feasibility for video restoration pipelines

## Visual Comparison

```python
img = Image.open("../data/test_frames/frame_00000.png")

print(img.size)

plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
original = Image.open("../data/test_frames/frame_00000.png")

restored = Image.open(
    "../data/test_restored_batch/Single_Image_Defocus_Deblurring/frame_00000.png"
)

fig, ax = plt.subplots(1, 2, figsize=(14, 6))

ax[0].imshow(original)
ax[0].set_title("Original")
ax[0].axis("off")

ax[1].imshow(restored)
ax[1].set_title("Restored")
ax[1].axis("off")

plt.tight_layout()
plt.show()

print("Original resolution:", original.size)
print("Restored resolution:", restored.size)
```

## Benchmark Results

```python
benchmark_results = pd.DataFrame({
    "Metric": [
        "Input Resolution",
        "Frames Processed",
        "Average Time per Frame",
        "Total Processing Time",
        "Hardware",
        "Model"
    ],
    "Value": [
        "1280 x 720",
        10,
        "29.7 sec",
        "4.95 min",
        "Apple M3 CPU",
        "Restormer"
    ]
})

benchmark_results
```

The benchmark was performed on an Apple Silicon MacBook Air M3 using the Restormer architecture for single-image defocus deblurring.

A total of 10 frames were processed to estimate average inference time and evaluate visual restoration quality.

## Conclusions

- Restormer successfully improved visual sharpness and reduced blur artifacts.
- Average inference time was approximately 29.7 seconds per frame.
- The model was executed locally without GPU acceleration.
- Results suggest that frame restoration is feasible on consumer hardware, although full-video processing remains computationally expensive.
- Future work includes testing additional restoration models and comparing performance across resolutions.
